```python
def get_title(soup):
  try:
      title = soup.find("h1", attrs={"class":'entry-title'}).text

  except AttributeError:
    title = ""

  return title

def get_body(soup):
  try:
      footer = soup.find('pre', class_='wp-block-preformatted')
      body = soup.find("div", attrs={"class":'td-ss-main-content'}).text.replace("\n","").replace("footer","")

  except AttributeError:
    body = ""

  return body
```

```python
d = {"Title":[], "Body":[]}

for i in df["URL"]:
  webpage = requests.get(i, headers=HEADERS)
  soup = BeautifulSoup(webpage.content, "html.parser")

  d['Title'].append(get_title(soup))
  d['Body'].append(get_body(soup))

n_df = pd.DataFrame.from_dict(d)
```

Percentage of Complex words

```python
def syllable_count(word):
    sc=0
    for w in word:
        if(w=='a' or w=='e' or w=='i' or w=='o' or w=='u' or w=='A' or w=='E' or w=='I' or w=='O' or w=='U'):
            sc+=1
    if word[-2:]=='es' and  word[-2:]=='ES' and  word[-2:]=='ed' and  word[-2:]=='ED':
        sc=sc-2
    return sc
```

```python
def complex_word_count(text):
    complex_word_count=0
    for word in text.split(" "):
        sc=syllable_count(word)
        if sc>=2:
            complex_word_count+=1
    return complex_word_count
```

```python
new_df['COMPLEX_WORD_COUNT']=new_df.text.apply(complex_word_count)
```

```python
new_df['PERCENTAGE OF COMPLEX WORDS']=(new_df['COMPLEX_WORD_COUNT']/new_df['WORD COUNT'])*100
```

```python
new_df
```

```python
def syllable_count_perword(text):
    syllable_count_list=[]
    for word in text.split(' '):
        if word in stop_words:
            continue
        else:
            sc=syllable_count(word)
            syllable_count_list.append(sc)

    scpw=np.mean(syllable_count_list)
    return scpw
```

```python
new_df['AVERAGE SYLLABLE COUNT PER WORD']=new_df.text.apply(syllable_count_perword)
```

```python
perpronoun=['I',"we","my","ours","us"]
```

```python
def perpronoun_count(text):
    pprc=0
    for word in text.split(' '):
        if word.lower() in perpronoun or word in perpronoun:#there's no such word as 'i' in Modern English.
            if word=='US':
                continue
            else:
                pprc+=1
    return pprc
```

```python
new_df['PERSONAL PRONOUNS']=new_df.text.apply(perpronoun_count)
```
